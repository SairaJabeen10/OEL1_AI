```python
from collections import deque

def find_reachable_shelters(roads, start):
    visited = set()
    queue = deque([start])
    reachable = []

    while queue:
        shelter = queue.popleft()
        if shelter not in visited:
            visited.add(shelter)
            reachable.append(shelter)
            queue.extend([neighbor for neighbor in roads.get(shelter, []) if neighbor not in visited])

    return reachable

roads = {
    'S1': ['S2', 'S3'],
    'S2': ['S1', 'S4'],
    'S3': ['S1', 'S5'],
    'S4': ['S2'],
    'S5': ['S3']
}

start_shelter = 'S1'
reachable_shelters = find_reachable_shelters(roads, start_shelter)

print(f"Reachable shelters from {start_shelter}: {reachable_shelters}")
```
